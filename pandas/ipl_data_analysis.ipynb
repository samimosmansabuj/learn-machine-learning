```python
import pandas as pd
```

```python
ipl_data = pd.read_csv("matches.csv")
ipl_data
```

***In which season were the most matches played?***

```python
ipl_data["season"].value_counts().sort_values(ascending=False).head(1)
```

```python
#using groupby
ipl_data.groupby("season").count()["id"].idxmax()
```

***Which venue hosted the most matches?***

```python
ipl_data["venue"].value_counts().idxmax()
```

```python
#using groupby
ipl_data.groupby("venue").count()["id"].sort_values(ascending=False).head(1)
```

***Which team has won the most matches?***

```python
# ipl_data["winner"]
# ipl_data.groupby("winner").count()["id"].sort_values(ascending=False).head(1)
matches_won = ipl_data["winner"].value_counts()
matches_won.head(1)
```

***Which team chose 'bat' the most after winning the toss?***

```python
chose_bat = ipl_data[ipl_data["toss_decision"] == "bat"]
# chose_bat["toss_winner"].value_counts().sort_values(ascending=False).head(1)
chose_bat["toss_winner"].value_counts().idxmax()
```

***Which player has won the most 'Player of the Match' awards?***

```python
player_of_match_counts = ipl_data["player_of_match"].value_counts().sort_values(ascending=False).head(1)
player_of_match_counts
```

***How many matches used the D/L method (Duckworth–Lewis)?***

```python
# ipl_data["dl_applied"].value_counts()
ipl_data[ipl_data["dl_applied"] == 1].count()["id"]
```
